```python
import pandas as pd
import pandas as np

policeIncidents = pd.read_csv("Incidents.csv") ## import data set
policeIncidents.head()
```

```python
#new dataframe with only the offense and neighborhood
main_df = policeIncidents[['INCIDENTNEIGHBORHOOD','OFFENSES']].copy()
main_df = main_df.sort_values('INCIDENTNEIGHBORHOOD')

main_df.head(10)
```

```python
#dict with neighborhood name & number of offenses (occurrences)
    
neighborhood_incidents = main_df.groupby('INCIDENTNEIGHBORHOOD').count().to_dict(orient='dict')['OFFENSES']
for n in neighborhood_incidents:
    print(n, neighborhood_incidents[n])
```
